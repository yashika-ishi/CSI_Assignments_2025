# **Step 1: Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Step 2: Load the Dataset**

```python
# Settings for better visuals
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Loading the Dataset
df = pd.read_csv('/content/drive/MyDrive/CSI DATASETS/Titanic_Dataset.csv')
print("First 5 rows of the dataset:")
print(df.head())
```

# **Step 3: Dataset Overview**

```python
print("\nDataset Shape:", df.shape)
print("\nDataset Info:")
df.info()
```

# **Step 4: Check for Missing Values**

```python
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
```

# **Step 5: Impute Missing Values**

```python
# Fill missing Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Fill missing Cabin with "Unknown"
df['Cabin'].fillna('Unknown', inplace=True)

# Verify again
print("\nMissing Values in Each Column (After Imputation):")
print(df.isnull().sum())
```

# **Step 6: Statistical Summary**

```python
print("\nSummary Statistics:")
print(df.describe(include='all'))
```

# **Step 7: Univariate Analysis**

```python
# Survival Count
sns.countplot(x='Survived', data=df, palette='pastel')
plt.title('Survival Distribution (0 = No, 1 = Yes)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

# Passenger Class
sns.countplot(x='Pclass', data=df, palette='cool')
plt.title('Passenger Class Distribution')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()

# Gender Distribution
sns.countplot(x='Sex', data=df, palette='Set2')
plt.title('Gender Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

# Age Distribution
sns.histplot(df['Age'].dropna(), bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Fare Distribution with Boxplot
sns.boxplot(y='Fare', data=df, color='lightgreen')
plt.title('Fare Boxplot (Detecting Outliers)')
plt.ylabel('Fare')
plt.show()
```

# **Step 8: Bivariate Analysis**

```python
# Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df, palette='coolwarm')
plt.title('Survival Count by Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()

# Survival by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set3')
plt.title('Survival Count by Class')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()

# Survival by Embarkation Port
sns.countplot(x='Embarked', hue='Survived', data=df, palette='Set1')
plt.title('Survival Count by Embarkation Port')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()

# Age vs Survival
sns.violinplot(x='Survived', y='Age', data=df, palette='muted')
plt.title('Age Distribution by Survival')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

# **Step 9: Correlation Analysis**

```python
# Compute correlation matrix
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# **Step 10: Feature Engineering**

```python
# Create FamilySize Feature
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Extract Title from Name
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

# Create Age Groups
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 20, 40, 60, 100],
                        labels=['Child', 'Teen', 'Adult', 'Middle-Aged', 'Senior'])
```

# **Step 11: Feature Visualizations**

```python
# Survival by Family Size
sns.countplot(x='FamilySize', hue='Survived', data=df, palette='viridis')
plt.title('Survival by Family Size')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()

# Survival by Title
sns.countplot(x='Title', hue='Survived', data=df, palette='Accent')
plt.title('Survival by Title')
plt.xlabel('Title')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Survived')
plt.show()

# Survival by Age Group
sns.countplot(x='AgeGroup', hue='Survived', data=df, palette='pastel')
plt.title('Survival by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()
```

# **Step 12: Outlier Detection**

```python
# Detecting Outliers in Age
sns.boxplot(y='Age', data=df, color='salmon')
plt.title('Age Outliers')
plt.ylabel('Age')
plt.show()

# Detecting Outliers in Fare
sns.boxplot(y='Fare', data=df, color='gold')
plt.title('Fare Outliers')
plt.ylabel('Fare')
plt.show()
```
